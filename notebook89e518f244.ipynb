```python
testing_data = pd.read_csv("/kaggle/input/home-data-for-ml-course/test.csv",index_col = 0)
training_data = pd.read_csv("/kaggle/input/home-data-for-ml-course/train.csv",index_col = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python

features = ["LotFrontage","LotArea","MiscVal","MSSubClass","ScreenPorch","MoSold","PoolArea"]
```

```python
for column in features:
    testing_data = testing_data.replace({column:{np.nan:testing_data[column].mean()}})
    training_data = training_data.replace({column:{np.nan:training_data[column].mean()}})
    
```

```python
training_data
```

```python
X_train,X_test,y_train,y_test = train_test_split(training_data[features], training_data['SalePrice'], test_size=0.2, random_state=42)
```

```python
model = RandomForestRegressor()
model.fit(X_train,y_train)
preds = model.predict(X_test)
print(mean_absolute_error(preds,y_test))
```
